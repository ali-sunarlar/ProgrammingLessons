```python
#TensorFlow ve Keras: Derin renme modelleri oluturmak iin kullanlr. 
#TensorFlow, Google tarafndan gelitirilmitir ve byk veri setleriyle alabilme yetenei ile bilinir. 
#Keras ise TensorFlow'un zerine ina edilmi, derin renme modellerini daha hzl ve kolay bir ekilde oluturmay salayan bir arayzdr. 
#Aada verilen kod sadece algoritmik anlama asndan verilmi olup kodun ilem detaylar ileri blmlerde ayrntl bir ekilde anlatlacaktr.

! pip install --trusted-host pypi.org --trusted-host files.pythonhosted.org tensorflow

! pip install --trusted-host pypi.org --trusted-host files.pythonhosted.org keras

import tensorflow as tf

# rnek veri

x_train = [1, 2, 3, 4]

y_train = [0, -1, -2, -3]

# Model parametreleri

W = tf.Variable([0.], dtype=tf.float32)

b = tf.Variable([0.], dtype=tf.float32)

# Modelin tanm

def linear_model(x):

return x * W + b

# Hata fonksiyonu (loss)

def loss(predicted_y, desired_y):

return tf.reduce_sum(tf.square(predicted_y - desired_y))

# Optimizasyon (SGD - Stochastic Gradient Descent)

optimizer = tf.optimizers.SGD(learning_rate=0.01)

# Eitim dngs

for i in range(1000):

with tf.GradientTape() as tape:

current_loss = loss(linear_model(x_train), y_train)

grads = tape.gradient(current_loss, [W, b])

optimizer.apply_gradients(zip(grads, [W, b]))

# Sonular ekrana yazdrma

print(f"W: {W.numpy()}, b: {b.numpy()}")
```

```python
#NLTK: Doal dil ileme (NLP) projeleri iin oluturulmu bir ktphanedir. 
#Metin madencilii, metin analizi ve dil ileme grevleri iin geni bir ara ve ngilizce dilindeki metinler iin byk miktarda veritaban ve kaynak ierir.
import nltk

from nltk.tokenize import word_tokenize

from nltk.probability import FreqDist

# rnek metin

text = "Doal dil ileme iin rnek metin girii buraya yazlr."

# Kelimelere ayrma (tokenization)

tokens = word_tokenize(text)

# Sklk dalmn hesaplama

fdist = FreqDist(tokens)

# En yaygn 5 kelimeyi yazdrma

print(fdist.most_common(5))
```

```python
#Aada verilen rnekte scikit-learn ktphanesi kullanlarak basit bir SVM uygulamasn gstermektedir. 
#Uygulamann karmaklna bal olarak bu ilem admlar geniletebilir veya zelletirebilir.
# Gerekli ktphanelerin yklenmesi

from sklearn import datasets

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC

# Veri setinin yklenmesi

iris = datasets.load_iris()

X = iris.data

y = iris.target

# Veri setinin eitim ve test olarak blnmesi

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Veri normalletirme

sc = StandardScaler()

X_train_std = sc.fit_transform(X_train)

X_test_std = sc.transform(X_test)

# SVM modelinin oluturulmas

svm = SVC(kernel='linear', C=1.0, random_state=42)

svm.fit(X_train_std, y_train)

# Modelin test verisi zerinde deerlendirilmesi

accuracy = svm.score(X_test_std, y_test)

print(f"Model Doruluu: {accuracy * 100:.2f}%")
```
